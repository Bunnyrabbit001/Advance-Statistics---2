**Q1. What is hypothesis testing in statistics?**
Hypothesis testing in statistics is a method used to make decisions or inferences about a population based on sample data. It helps determine whether there is enough evidence in a sample to support a particular belief (hypothesis) about a population parameter.

**Key Terms:**

1. **Null Hypothesis (H‚ÇÄ):**  
   The default or original assumption. Usually, it states that there is no effect or no difference.  
   Example: *H‚ÇÄ: The average height of students is 5.5 feet.*

2. **Alternative Hypothesis (H‚ÇÅ or Ha):**  
   The statement you want to test or prove. It‚Äôs what you conclude if you reject the null.  
   Example: *H‚ÇÅ: The average height of students is not 5.5 feet.*

3. **Test Statistic:**  
   A standardized value calculated from sample data that is used in making the decision. Examples include the Z-statistic, t-statistic, etc.

4. **Significance Level (Œ±):**  
   The probability of rejecting the null hypothesis when it is actually true (Type I error). Common values are 0.05 or 0.01.

5. **P-value:**  
   The probability of observing the test results (or more extreme) under the assumption that the null hypothesis is true.

6. **Decision Rule:**  
   - If **p-value ‚â§ Œ±**, reject H‚ÇÄ (there is enough evidence).
   - If **p-value > Œ±**, fail to reject H‚ÇÄ (not enough evidence).



**Q2.What is the null hypothesis, and how does it differ from the alternative hypothesis**

The **null hypothesis (H‚ÇÄ)** and the **alternative hypothesis (H‚ÇÅ or Ha)** are two opposing statements used in hypothesis testing:


**1. Null Hypothesis (H‚ÇÄ):**

- It represents the **default** or **status quo** assumption.
- It suggests **no effect**, **no difference**, or **no relationship**.
- The goal of hypothesis testing is to **test** this assumption using sample data.

**Example:**  
H‚ÇÄ: The average salary of employees is ‚Çπ50,000.  
(This means there‚Äôs no change or deviation from ‚Çπ50,000.)

---

**2. Alternative Hypothesis (H‚ÇÅ or Ha):**

- It‚Äôs what you want to **test for** or **prove**.
- It suggests **a change**, **a difference**, or **an effect**.
- If there‚Äôs enough evidence against H‚ÇÄ, you **accept H‚ÇÅ**.

**Example:**  
H‚ÇÅ: The average salary of employees is **not** ‚Çπ50,000.  
(This means there‚Äôs some change‚Äîeither more or less.)

---

**Key Difference:**

| Feature               | Null Hypothesis (H‚ÇÄ)             | Alternative Hypothesis (H‚ÇÅ / Ha)     |
|----------------------|----------------------------------|--------------------------------------|
| Meaning              | No effect or change              | There is an effect or change         |
| Role in testing      | Assumed true until proven wrong  | Opposed to H‚ÇÄ, tested as a challenger |
| Rejected when        | Evidence strongly contradicts it | Evidence supports this view          |

---


**Q3. What is the significance level in hypothesis testing, and why is it important?**

The **significance level** (denoted as **Œ±**) is a threshold used in hypothesis testing to decide whether to reject the null hypothesis (**H‚ÇÄ**). It represents the **maximum probability of making a Type I error**, which means **rejecting H‚ÇÄ when it is actually true**.

---

**Why is it important?**

- It sets the **cut-off** for determining how **extreme** your results need to be before you say they are statistically significant.
- It **controls the risk** of making a false positive decision (i.e., claiming a finding when there's none).

---

**Common Significance Levels:**

| Œ± value | Interpretation                     |
|---------|------------------------------------|
| 0.05    | 5% chance of rejecting a true H‚ÇÄ   |
| 0.01    | 1% chance ‚Äî more strict            |
| 0.10    | 10% chance ‚Äî more lenient          |

---

 **How it's used in decision making:**

- If the **p-value ‚â§ Œ±** ‚Üí **Reject H‚ÇÄ** ‚Üí The result is **statistically significant**.
- If the **p-value > Œ±** ‚Üí **Fail to reject H‚ÇÄ** ‚Üí The result is **not statistically significant**.

---

**Example:**

Suppose you're testing whether a new medicine is more effective than the standard one.

- **H‚ÇÄ**: There is no difference in effectiveness.
- **H‚ÇÅ**: The new medicine is more effective.
- You choose **Œ± = 0.05** (5%).



**Q4. What does a P-value represent in hypothesis testing?**

The **p-value** in hypothesis testing represents the **probability of obtaining a result as extreme as (or more extreme than) the one observed**, **assuming the null hypothesis (H‚ÇÄ) is true**.

---

**In simple terms:**

It answers the question:  
**"If H‚ÇÄ is true, how likely is it that I would get this sample result just by chance?"**

---

**How to interpret the p-value:**

| **P-value**            | **Interpretation**                                              |
|------------------------|------------------------------------------------------------------|
| **Small p-value (‚â§ Œ±)**| Strong evidence **against** H‚ÇÄ ‚Üí **Reject H‚ÇÄ**                  |
| **Large p-value (> Œ±)**| Weak evidence against H‚ÇÄ ‚Üí **Fail to reject H‚ÇÄ**                |

---

**Example:**

- **H‚ÇÄ**: A coin is fair.
- You flip the coin 100 times and get 80 heads.
- The p-value tells you how likely it is to get 80 or more heads if the coin were actually fair.

If the **p-value = 0.01**, there‚Äôs a **1% chance** of getting such a result **by random chance**. That‚Äôs quite low, so you‚Äôd **reject the idea that the coin is fair**.

**Q6. What are Type 1 and Type 2 errors in hypothesis testing?**
In hypothesis testing, **Type I and Type II errors** represent the two possible ways we can make **wrong decisions** based on sample data.


**Type I Error (False Positive):**
- You conclude that a difference or effect exists **when it actually doesn't**.
- Controlled by the **significance level Œ±** (e.g., 0.05 means 5% risk of making this error).

**Example:**  
You say a new drug works better than the old one, but in reality, it doesn‚Äôt.

---

**Type II Error (False Negative):**
- You **miss** a real effect.
- Represented by **Œ≤** (beta), and the chance of avoiding it is called **power** (1 - Œ≤).

**Example:**  
You say a new drug has no effect, but it actually works.

- **H‚ÇÄ:** New = Old  
- **H‚ÇÅ:** New ‚â† Old  
‚Üí This test checks both **better** and **worse** outcomes.

**Use when:** You want to detect **any significant change**, regardless of direction.




**Q8.What is the Z-test, and when is it used in hypothesis testing**

The **Z-test** is a type of **hypothesis test** used to determine whether there is a significant difference between sample and population parameters, or between two samples, **when the population variance is known** and/or the sample size is large (typically **n ‚â• 30**).

 **When to Use a Z-test:**

1. **The sample size is large** (n ‚â• 30).
2. **Population standard deviation (œÉ) is known**.
3. The data is **normally distributed** or approximately normal due to the Central Limit Theorem.


**Types of Z-tests:**

| Type of Z-test                 | When It's Used                                               |
|-------------------------------|---------------------------------------------------------------|
| **One-sample Z-test**         | Compare sample mean to known population mean                 |
| **Two-sample Z-test**         | Compare means of two independent large samples               |
| **Z-test for proportions**    | Compare sample proportion to population or between two samples |

---

**Example (One-sample Z-test):**

Suppose the average weight of bags produced by a machine is **50 kg** (œÉ = 5 kg). You take a sample of 40 bags and find a sample mean of 52 kg.

- **H‚ÇÄ:** Œº = 50  
- **H‚ÇÅ:** Œº ‚â† 50  
- Use the Z-test to check if the difference is significant.

---

**Z-test Formula (for one sample mean):**

\\begin{cases}
[
Z = \frac{\bar{X} - \mu}{\sigma / \sqrt{n}}
\end{cases}

Where:  
- \(\bar{X}\) = sample mean  
- \(\mu\) = population mean  
- \(\sigma\) = population standard deviation  
- \(n\) = sample size

**Q9.  How do you calculate the Z-score, and what does it represent in hypothesis testing?**

**How to Calculate the Z-score in Hypothesis Testing:**

The **Z-score** measures how many **standard errors** a sample statistic is from the **population parameter** under the **null hypothesis (H‚ÇÄ)**.

---

**Formula (for sample mean):**

\begin{cases}
[
Z = \frac{\bar{X} - \mu}{\sigma / \sqrt{n}}
]
\end{cases}

Where:  
- \begin{cases} (\bar{X}\) = sample mean \end{cases}  
- \begin{cases} (\mu\) = population mean (under H‚ÇÄ)  \end{cases}
- \begin{cases} (\sigma\) = population standard deviation  \end{cases}
- \begin{cases} (n\) = sample size  \end{cases}

---

**What the Z-score Represents:**

- It tells you **how far your sample result is from the assumed population mean**, in units of standard error.
- A **large positive or negative Z-score** indicates that the sample result is **far from the population mean**, and may not have occurred by random chance.

---

**In Hypothesis Testing:**

- You compare the Z-score to **critical values** (like ¬±1.96 for Œ± = 0.05 in a two-tailed test).
- Or you convert the Z-score to a **p-value**, and:
  - If **p-value ‚â§ Œ±** ‚Üí **Reject H‚ÇÄ**
  - If **p-value > Œ±** ‚Üí **Fail to reject H‚ÇÄ**

---

**Example:**

Let‚Äôs say:  
- Population mean (Œº) = 100  
- Population standard deviation (œÉ) = 10  
- Sample mean (\(\bar{X}\)) = 104  
- Sample size (n) = 25  

Then:

\begin{cases}
[
Z = \frac{104 - 100}{10 / \sqrt{25}} = \frac{4}{2} = 2.0
]
\end{cases}

A Z-score of 2.0 tells you the sample mean is 2 standard errors **above** the population mean.

**Q12. What is the T-test, and how is it used in hypothesis testing?**

**Example:**

 **Q14. What is a confidence interval, and how is it used to interpret statistical results**


A **confidence interval** is a **range of values** that is likely to contain the **true population parameter** (like the mean or proportion) with a certain level of confidence ‚Äî usually **95%** or **99%**.

It gives an estimated range based on your **sample data**, showing how much **uncertainty** there is around your estimate.


**General Formula (for mean):**

\begin{cases}
\text{Confidence Interval} = \bar{X} \pm ( \text{critical value} \times \text{standard error} )
\end{cases}

Where:  
- \(\bar{X}\) = sample mean  
- **Critical value** = depends on confidence level (e.g., 1.96 for 95% in Z, or a t-score for small samples)  
- **Standard error** = \begin{cases} (\frac{s}{\sqrt{n}}\) or \(\frac{\sigma}{\sqrt{n}}\end{cases})


**How It‚Äôs Used to Interpret Results:**

- A **95% confidence interval** means:  
  ‚ÄúWe are 95% confident that the true population value lies **within this interval**.‚Äù

- If you're testing a **hypothesis about a mean**:
  - If the **null hypothesis value (e.g., Œº = 0)** is **inside** the CI ‚Üí **Do not reject H‚ÇÄ**.
  - If it's **outside** the CI ‚Üí **Reject H‚ÇÄ**.



**Example:**

Let‚Äôs say:
- Sample mean = 70  
- Standard deviation = 10  
- Sample size = 25  
- Confidence level = 95%  
- Standard error = 10 / ‚àö25 = 2  
- Critical Z = 1.96 (for 95% confidence)

\begin{cases}
\text{CI} = 70 \pm (1.96 \times 2) = 70 \pm 3.92
\end{cases}

\begin{cases}
\Rightarrow (66.08, 73.92)
\end{cases}

> We‚Äôre 95% confident that the **true population mean** lies between **66.08 and 73.92**.



**Q16.  How is Bayes' Theorem used in statistics, and what is its significance.**

### üîπ **What is Bayes' Theorem?**

**Bayes' Theorem** is a mathematical formula used in statistics to **update probabilities** based on new evidence.  
It answers the question:  
> ‚ÄúGiven that we observed some result, what is the probability that a particular cause or condition is true?‚Äù

---

### ‚úÖ **Bayes' Theorem Formula:**

\begin{cases}
[
P(A \mid B) = \frac{P(B \mid A) \cdot P(A)}{P(B)}
]
\end{cases}


Where:
- \(P(A \mid B)\) = **Posterior** probability (probability of A given B)
- \(P(B \mid A)\) = **Likelihood** (probability of observing B given A)
- \(P(A)\) = **Prior** probability (initial belief about A)
- \(P(B)\) = **Evidence** or total probability of B

---

**Significance in Statistics:**

1. **Incorporates prior knowledge**: Bayes‚Äô theorem lets you combine previous beliefs (priors) with new data (evidence).
2. **Used in Bayesian statistics**: Helps make **probability-based decisions** and **model uncertainty**.
3. **Foundation of machine learning & AI models**: Especially in **Naive Bayes classifiers**, **Bayesian networks**, etc.
4. **Improves decision-making under uncertainty**.

---

**Real-World Example (Medical Testing):**

Suppose:
- A disease affects 1 in 1,000 people: \(P(D) = 0.001\)
- A test is 99% accurate:  
  - \begin{cases}(P(\text{Positive} \mid D) = 0.99\end{cases})  
  - \begin{cases}(P(\text{Positive} \mid \text{No Disease}) = 0.01\end{cases})

You test positive. What's the chance you actually have the disease?  
Use Bayes‚Äô Theorem to find \begin{cases}(P(D \mid \text{Positive})\end{cases}).  
Result: **It's much lower than 99%** ‚Äî because the disease is rare.


**Q17.What is the Chi-square distribution, and when is it used?**


The **Chi-square (œá¬≤) distribution** is a **right-skewed probability distribution** used primarily for **categorical data**. It is based on the **sum of squared standard normal variables**, and its shape depends on **degrees of freedom (df)**.


**Key Characteristics:**

- Only **non-negative** values (œá¬≤ ‚â• 0)
- **Skewed** to the right, especially with small df
- As df increases, it becomes more **normal-like**

---

**When Is the Chi-Square Distribution Used?**

It‚Äôs used in **hypothesis testing** for:
1. **Goodness-of-Fit Test**  
   - To check if a sample matches a **theoretical distribution**  
   - Example: Are dice rolls uniformly distributed?

2. **Test of Independence** (Chi-square test in a contingency table)  
   - To see if **two categorical variables are related**  
   - Example: Is gender independent of voting preference?

3. **Test for Homogeneity**  
   - To compare **distributions across different groups**

4. **Confidence intervals or hypothesis tests** about **variance or standard deviation** (less common)

**Chi-Square Test Statistic Formula:**

\begin{cases}
[
\chi^2 = \sum \frac{(O_i - E_i)^2}{E_i}
]
\end{cases}

Where:  
- \(O_i\) = observed frequency  
- \(E_i\) = expected frequency

---

**Example (Independence Test):**

|           | Likes Tea | Likes Coffee | Total |
|-----------|-----------|--------------|-------|
| Male      |    30     |     20       |  50   |
| Female    |    10     |     40       |  50   |
| **Total** |    40     |     60       | 100   |

- Null Hypothesis (H‚ÇÄ): Preference is **independent** of gender.
- Use a **Chi-square test** to check if there's a **significant relationship**.


**Q18. What is the Chi-square goodness of fit test, and how is it applied?**


The **Chi-square goodness of fit test** is a statistical test used to determine **whether a sample data set matches an expected (theoretical) distribution**.

It helps answer questions like:  
> ‚ÄúDoes the observed frequency distribution of a categorical variable fit the expected distribution?‚Äù


**When to Use It:**

- You have **categorical data**.
- You know the **expected proportions or frequencies**.
- You want to test **how well your observed data fits those expectations**.

---

**Test Statistic Formula:**

\begin{cases}
[
\chi^2 = \sum \frac{(O_i - E_i)^2}{E_i}
]
\end{cases}

Where:  
- \(O_i\) = observed frequency for category \(i\)  
- \(E_i\) = expected frequency for category \(i\)  
- The sum runs over all categories.


**Steps to Apply the Test:**

1. **State Hypotheses:**
   - **H‚ÇÄ**: The observed distribution **matches** the expected distribution.
   - **H‚ÇÅ**: The observed distribution **does not match** the expected distribution.

2. **Calculate Expected Frequencies**  
   Based on given proportions or assumptions (e.g., uniform distribution).

3. **Compute the Chi-square statistic** using the formula above.

4. **Find the critical value or p-value** using the **œá¬≤-distribution** with  
   \begin {cases}
   [
   \text{df} = \text{number of categories} - 1
   ]
   \end{cases}

5. **Decision Rule:**
   - If **œá¬≤ > critical value**, or **p-value ‚â§ Œ±**, ‚Üí **Reject H‚ÇÄ**.


**Example:**

You roll a die 60 times. Here are your results:

| Face | Observed (O) | Expected (E = 60/6 = 10) |
|------|--------------|--------------------------|
| 1    |     8        |           10             |
| 2    |    10        |           10             |
| 3    |    12        |           10             |
| 4    |     9        |           10             |
| 5    |    11        |           10             |
| 6    |    10        |           10             |

\begin{cases}
[
\chi^2 = \sum \frac{(O - E)^2}{E} = \frac{(8-10)^2}{10} + \cdots + \frac{(10-10)^2}{10} = 1.0
]
\end{cases}

- df = 6 ‚àí 1 = 5
- If the critical value at Œ± = 0.05 is 11.07 ‚Üí **Since 1.0 < 11.07**, we **fail to reject H‚ÇÄ**.



**Q19.  What is the F-distribution, and when is it used in hypothesis testing**

**What is the F-Distribution?**

The **F-distribution** is a **right-skewed probability distribution** used mainly to **compare variances** between two populations. It‚Äôs formed by taking the ratio of two independent chi-square variables (each divided by its degrees of freedom).

---

**Key Properties:**

- **Only positive values** (F ‚â• 0)
- Shape depends on **two sets of degrees of freedom**: \(df_1\) (numerator), \(df_2\) (denominator)
- Used when dealing with **variability**, especially **ratios of variances**

---

**When is the F-Distribution Used?**

1. **ANOVA (Analysis of Variance)**  
   - To test whether **3 or more group means are significantly different**  
   - Compares **variances between groups vs. within groups**

2. **Comparing Two Variances**  
   - F-test for equality of variances  
   - Example: Is the variance in test scores higher in one class than another?

3. **Regression Analysis**  
   - Tests whether your regression model as a whole is statistically significant

---

**F-Test Statistic Formula (for variance comparison):**

\begin {cases}
[
F = \frac{s_1^2}{s_2^2}
]
\end{cases}

Where:
- \(s_1^2\) = sample variance of group 1  
- \(s_2^2\) = sample variance of group 2  
- Typically, put the **larger variance** on top to keep \(F \geq 1\)

---

**Q20.What is an ANOVA test, and what are its assumptions**

**ANOVA** (Analysis of Variance) is a statistical method used to **compare the means of three or more groups** to see if at least one of them is **significantly different** from the others.

> It tests the hypothesis:  
> H‚ÇÄ: All group means are equal  
> H‚ÇÅ: At least one group mean is different

**When to Use ANOVA:**

- You have **3 or more independent groups**.
- Your **dependent variable is continuous** (like height, test scores, etc.).
- You want to test **differences in means**, not variances.

**How It Works:**

ANOVA compares:
- **Between-group variability** (how far group means are from the overall mean)
- **Within-group variability** (how much individuals in a group vary from their group mean)

It calculates an **F-statistic**:

\begin{cases}
[
F = \frac{\text{Between-group variance}}{\text{Within-group variance}}
]
\end{cases}

If the **F-value is large**, it suggests group means are not all the same.



**Example:**
You test 3 teaching methods on student scores. ANOVA tells you whether **any of the methods** lead to **different average scores**, but not *which ones* ‚Äî you'd need **post hoc tests** (like Tukey‚Äôs) for that.



**Key Assumptions of ANOVA:**

1. **Independence of observations**  
   - Data from each group should be collected independently.

2. **Normality**  
   - The **dependent variable** should be **approximately normally distributed** within each group.

3. **Homogeneity of variances** (Equal variance)  
   - The **variance in each group should be roughly equal**. This is often tested using **Levene‚Äôs test**.

---

**If Assumptions Are Violated:**

- **Mild violations of normality**: ANOVA is robust.  
- **Violation of equal variances**: Use **Welch‚Äôs ANOVA**.  
- **Non-normal data or ordinal data**: Use **Kruskal-Wallis test** instead.


**Q21. What are the different types of ANOVA tests?**

ANOVA isn't a one-size-fits-all test ‚Äî there are **different types**, each suited to specific experimental designs or data structures. Here are the **main types of ANOVA tests**:

---
1. **One-Way ANOVA**  
**Use when:** Comparing the means of **3 or more independent groups** based on **one factor**.

- **Example:** Compare test scores among students taught by 3 different teachers.  
- **Groups are unrelated**, and you're testing **one independent variable** (e.g., teaching method).

---

2. **Two-Way ANOVA (Factorial ANOVA)**  
**Use when:** Comparing means based on **two independent variables (factors)**.

- **Example:** Study the effect of **teaching method** and **study time** on test scores.  
- Tests:
  - Main effect of factor A
  - Main effect of factor B
  - **Interaction effect** between A and B (whether the effect of one factor depends on the other)

---
3. **Repeated Measures ANOVA**  
 **Use when:** The **same subjects** are measured **multiple times** under different conditions or time points.

- **Example:** Measuring blood pressure of patients **before, during, and after** treatment.
- It accounts for **correlation between repeated measurements** of the same individual.

---

4. **Mixed-Design ANOVA (Split-Plot ANOVA)**  
**Use when:** You have both **between-subjects** and **within-subjects** variables.

- **Example:** Comparing 2 groups (between-subjects: treatment vs. control) over 3 time points (within-subjects: pre-, mid-, post-test).
- It blends **repeated measures** with **independent group comparisons**.


5. **MANOVA (Multivariate Analysis of Variance)**  
**Use when:** You have **multiple dependent variables** and want to see how groups differ across them **simultaneously**.

- **Example:** Measuring the impact of diet on **weight**, **cholesterol**, and **blood pressure** together.

---


**Q22. What is the F-test, and how does it relate to hypothesis testing**

The **F-test** is a statistical test used to **compare two variances** or to **analyze multiple means** (as in ANOVA). It relies on the **F-distribution** and helps determine whether **variability** between groups is statistically significant.

---

**Common Uses of the F-Test:**

1. **Comparing Two Variances**  
   - Tests if the **variances** of two populations are equal.  
   - Example: Is the spread of test scores in Class A different from Class B?

2. **In ANOVA (Analysis of Variance)**  
   - The F-test checks if **group means** differ significantly by comparing **between-group variance** to **within-group variance**.

3. **In Regression Analysis**  
   - Tests whether a **regression model** explains a significant portion of the variation in the dependent variable.

**F-Test Statistic Formula (Variance Comparison):**

\begin{cases}
[
F = \frac{s_1^2}{s_2^2}
]
\end{cases}

Where:
- \(s_1^2\), \(s_2^2\) = sample variances  
- Usually, \(s_1^2\) is the larger variance so that \(F \geq 1\)

For **ANOVA**:

\begin{cases}
[
F = \frac{\text{Mean Square Between (MSB)}}{\text{Mean Square Within (MSW)}}
]
\end{cases}


**How It Relates to Hypothesis Testing:**

- **Null Hypothesis (H‚ÇÄ):**
  - In variance comparison: The variances are equal.
  - In ANOVA: All group means are equal.
  - In regression: The model has **no explanatory power** (all coefficients = 0).

- **Alternative Hypothesis (H‚ÇÅ):**
  - At least one variance (or mean) is different.

- **Decision Rule:**
  - If the calculated F-value is **greater than the critical value**, or  
  - If the **p-value is less than Œ± (e.g., 0.05)**  
  ‚Üí **Reject H‚ÇÄ** (suggests significant difference).

```python
#Q1. Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results.

import numpy as np
from scipy import stats

# Generate random sample data
np.random.seed(42)  # for reproducibility
sample = np.random.normal(loc=105, scale=15, size=30)  # mean=105, std=15, n=30

# Known population parameters
population_mean = 100
population_std = 15  # Known population standard deviation

# Calculate sample statistics
sample_mean = np.mean(sample)
sample_size = len(sample)

# Calculate the Z-score
z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

# Calculate the p-value (two-tailed test)
p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

# Print results
print("Sample Mean:", round(sample_mean, 2))
print("Z-Score:", round(z_score, 4))
print("P-Value:", round(p_value, 4))

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Result: Reject the null hypothesis ‚Äî significant difference from the population mean.")
else:
    print("Result: Fail to reject the null hypothesis ‚Äî no significant difference from the population mean.")
```

```python
#Q5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def visualize_type1_type2(mu0=100, mu1=105, sigma=10, n=30, alpha=0.05):

    # Standard error
    se = sigma / np.sqrt(n)

    # Critical value for right-tailed test
    z_critical = norm.ppf(1 - alpha)
    x_crit = mu0 + z_critical * se

    # Beta: Type II error (probability of accepting H0 when H1 is true)
    z_beta = (x_crit - mu1) / se
    beta = norm.cdf(z_beta)

    # X range for plotting
    x = np.linspace(mu0 - 4*se, mu1 + 4*se, 1000)

    # Null and alternative distributions
    y_h0 = norm.pdf(x, mu0, se)
    y_h1 = norm.pdf(x, mu1, se)

    # Plotting
    plt.figure(figsize=(10, 6))
    plt.plot(x, y_h0, label='H‚ÇÄ Distribution', color='blue')
    plt.plot(x, y_h1, label='H‚ÇÅ Distribution', color='green')

    # Type I Error (alpha) region
    plt.fill_between(x, y_h0, where=(x >= x_crit), color='red', alpha=0.4, label='Type I Error (Œ±)')

    # Type II Error (beta) region
    plt.fill_between(x, y_h1, where=(x < x_crit), color='orange', alpha=0.4, label='Type II Error (Œ≤)')

    # Lines and annotations
    plt.axvline(x_crit, color='black', linestyle='--', label=f'Critical Value = {x_crit:.2f}')
    plt.title("Type I and Type II Errors Visualization")
    plt.xlabel("Sample Mean")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.grid(True)
    plt.show()

    # Print error probabilities
    print(f"Critical Value: {x_crit:.2f}")
    print(f"Type I Error (Œ±): {alpha}")
    print(f"Type II Error (Œ≤): {beta:.4f}")
    print(f"Power of the test (1 - Œ≤): {1 - beta:.4f}")

# Example usage
visualize_type1_type2(mu0=100, mu1=105, sigma=10, n=30, alpha=0.05)
```

```python
#Q9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance.

import numpy as np
from scipy import stats

def confidence_interval(data, confidence=0.95):
    """
    Calculates the confidence interval for the mean of a sample.

    Parameters:
        data (array-like): Sample data
        confidence (float): Confidence level (default 95%)

    Returns:
        (lower_bound, upper_bound)
    """
    data = np.array(data)
    n = len(data)
    mean = np.mean(data)
    sem = stats.sem(data)  # Standard error of the mean
    margin = sem * stats.t.ppf((1 + confidence) / 2, df=n - 1)
    return mean - margin, mean + margin

# Example usage
np.random.seed(42)
sample = np.random.normal(loc=75, scale=10, size=40)  # Simulated sample
ci = confidence_interval(sample, confidence=0.95)

print(f"95% Confidence Interval for the Sample Mean: ({ci[0]:.2f}, {ci[1]:.2f})")

```

```python
#Q10.Write a Python program to calculate the margin of error for a given confidence level using sample data.

import numpy as np
from scipy import stats

def margin_of_error(data, confidence=0.95):
    """
    Calculates the margin of error for a given confidence level using sample data.

    Parameters:
        data (array-like): The sample data
        confidence (float): The confidence level (default is 0.95 for 95%)

    Returns:
        float: Margin of error
    """
    data = np.array(data)
    n = len(data)
    sem = stats.sem(data)  # Standard error of the mean
    t_critical = stats.t.ppf((1 + confidence) / 2, df=n - 1)
    moe = t_critical * sem
    return moe

# Example usage
np.random.seed(0)
sample = np.random.normal(loc=50, scale=8, size=40)

moe_95 = margin_of_error(sample, confidence=0.95)

print(f"Margin of Error at 95% Confidence: ¬±{moe_95:.2f}")
```

```python
#Q11. Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.
def bayes_theorem(prior_A, likelihood_B_given_A, likelihood_B_given_not_A):

    prior_not_A = 1 - prior_A
    numerator = likelihood_B_given_A * prior_A
    denominator = (likelihood_B_given_A * prior_A) + (likelihood_B_given_not_A * prior_not_A)
    posterior = numerator / denominator
    return posterior

# Example scenario
prior = 0.01                      # P(Disease)
sensitivity = 0.95               # P(Positive | Disease)
false_positive_rate = 0.05       # P(Positive | No Disease)

posterior = bayes_theorem(prior, sensitivity, false_positive_rate)

print(f"Probability of having the disease given a positive test: {posterior:.4f} ({posterior*100:.2f}%)")
```

```python
#Q16.Implement an F-test using Python to compare the variances of two random samples.

import numpy as np
from scipy.stats import f

def f_test(sample1, sample2, alpha=0.05):
    """
    Performs an F-test to compare the variances of two samples.

    Parameters:
        sample1, sample2: Arrays of sample data
        alpha: Significance level (default 0.05)

    Returns:
        F-statistic, p-value, conclusion
    """
    var1 = np.var(sample1, ddof=1)
    var2 = np.var(sample2, ddof=1)

    # Ensure F >= 1
    if var1 > var2:
        f_stat = var1 / var2
        dfn, dfd = len(sample1) - 1, len(sample2) - 1
    else:
        f_stat = var2 / var1
        dfn, dfd = len(sample2) - 1, len(sample1) - 1

    p_value = 2 * min(f.cdf(f_stat, dfn, dfd), 1 - f.cdf(f_stat, dfn, dfd))

    conclusion = "Reject H‚ÇÄ: Variances are significantly different" if p_value < alpha else "Fail to reject H‚ÇÄ: No significant difference in variances"

    return f_stat, p_value, conclusion

# Example: Simulate data
np.random.seed(1)
sample1 = np.random.normal(loc=50, scale=10, size=30)
sample2 = np.random.normal(loc=50, scale=15, size=30)

# Perform F-test
f_stat, p_val, result = f_test(sample1, sample2)

print(f"F-statistic: {f_stat:.4f}")
print(f"P-value: {p_val:.4f}")
print("Conclusion:", result)
```

```python
#Q17.Write a Python program to perform an ANOVA test to compare means between multiple groups and interpret the results.

import numpy as np
from scipy.stats import f_oneway

# Step 1: Simulate sample data for 3 groups (e.g., test scores of 3 teaching methods)
np.random.seed(42)
group1 = np.random.normal(loc=75, scale=5, size=30)  # Group 1
group2 = np.random.normal(loc=78, scale=5, size=30)  # Group 2
group3 = np.random.normal(loc=72, scale=5, size=30)  # Group 3

# Step 2: Perform one-way ANOVA test
f_stat, p_value = f_oneway(group1, group2, group3)

# Step 3: Display results
print("ANOVA Test Results")
print(f"F-statistic: {f_stat:.4f}")
print(f"P-value: {p_value:.4f}")

# Step 4: Interpretation
alpha = 0.05
if p_value < alpha:
    print("Conclusion: Reject H‚ÇÄ ‚Äî At least one group mean is significantly different.")
else:
    print("Conclusion: Fail to reject H‚ÇÄ ‚Äî No significant difference between group means.")
```

```python
#Q18.  Perform a one-way ANOVA test using Python to compare the means of different groups and plot the results

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import f_oneway

# Step 1: Simulate sample data for three groups
np.random.seed(0)
group_A = np.random.normal(loc=70, scale=5, size=30)
group_B = np.random.normal(loc=75, scale=5, size=30)
group_C = np.random.normal(loc=80, scale=5, size=30)

# Step 2: Combine into a DataFrame
df = pd.DataFrame({
    'Score': np.concatenate([group_A, group_B, group_C]),
    'Group': ['A']*30 + ['B']*30 + ['C']*30
})

# Step 3: Perform One-Way ANOVA
f_stat, p_val = f_oneway(group_A, group_B, group_C)

# Step 4: Output results
print("One-Way ANOVA Results")
print(f"F-statistic: {f_stat:.4f}")
print(f"P-value: {p_val:.4f}")

if p_val < 0.05:
    print("Conclusion: Reject H‚ÇÄ ‚Äî At least one group mean is significantly different.")
else:
    print("Conclusion: Fail to reject H‚ÇÄ ‚Äî No significant difference between group means.")

# Step 5: Visualize with boxplot
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x='Group', y='Score', palette='Set2')
plt.title('Group Comparison using One-Way ANOVA')
plt.ylabel('Score')
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

```python
#Q19.Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA.

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import shapiro, levene
from statsmodels.stats.diagnostic import lilliefors

def check_anova_assumptions(data, group_col, value_col):
    groups = data[group_col].unique()

    print("üîç Checking ANOVA Assumptions\n")

    # 1. Normality: Shapiro-Wilk test per group
    print("1. Normality (Shapiro-Wilk test):")
    for group in groups:
        stat, p = shapiro(data[data[group_col] == group][value_col])
        print(f"  Group {group}: p-value = {p:.4f} {' Normal' if p > 0.05 else 'Not Normal'}")

    # 2. Homogeneity of variances: Levene's Test
    print("\n2. Homogeneity of Variances (Levene‚Äôs test):")
    group_values = [data[data[group_col] == g][value_col] for g in groups]
    stat, p = levene(*group_values)
    print(f"  Levene‚Äôs test p-value = {p:.4f} {'Equal variances' if p > 0.05 else 'Variances differ'}")

    # 3. Independence: Can‚Äôt be tested statistically, assumed through proper data collection
    print("\n3. Independence:")
    print("  Independence must be ensured by study design (random sampling, no repeated measures).")

    # Optional: Visual check
    plt.figure(figsize=(10, 4))
    plt.subplot(1, 2, 1)
    sns.boxplot(x=group_col, y=value_col, data=data)
    plt.title("Boxplot by Group")

    plt.subplot(1, 2, 2)
    sns.histplot(data=value_col, kde=True, bins=20)
    plt.title("Overall Histogram of Values")

    plt.tight_layout()
    plt.show()

# Simulate some sample data
np.random.seed(1)
df = pd.DataFrame({
    'Group': ['A']*30 + ['B']*30 + ['C']*30,
    'Score': np.concatenate([
        np.random.normal(70, 5, 30),
        np.random.normal(72, 5, 30),
        np.random.normal(75, 6, 30)
    ])
})

check_anova_assumptions(df, group_col='Group', value_col='Score')

```

```python
#Q20.Perform a two-way ANOVA test using Python to study the interaction between two factors and visualize the results.

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
from statsmodels.formula.api import ols
from statsmodels.graphics.factorplots import interaction_plot

# Step 1: Simulate sample data
np.random.seed(42)

# Factor A: Teaching Method (A1, A2)
# Factor B: School (S1, S2, S3)
teaching_method = np.repeat(['A1', 'A2'], 30)
school = np.tile(np.repeat(['S1', 'S2', 'S3'], 10), 2)

# Response variable: Student score with some interaction effect
scores = (np.random.normal(70, 5, 60) +
          (teaching_method == 'A2') * 3 +
          (school == 'S2') * 2 +
          ((teaching_method == 'A2') & (school == 'S3')) * 2)

# Create DataFrame
df = pd.DataFrame({
    'Score': scores,
    'TeachingMethod': teaching_method,
    'School': school
})

# Step 2: Perform Two-Way ANOVA
model = ols('Score ~ C(TeachingMethod) + C(School) + C(TeachingMethod):C(School)', data=df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)

# Step 3: Print ANOVA table
print("Two-Way ANOVA Results")
print(anova_table)

# Step 4: Visualize interaction effect
plt.figure(figsize=(8, 6))
interaction_plot(df['School'], df['TeachingMethod'], df['Score'],
                 colors=['red', 'blue'], markers=['D', '^'], ms=10)
plt.title("Interaction Plot: Teaching Method vs School")
plt.ylabel("Mean Score")
plt.grid(True)
plt.show()
```

```python
#Q21.  Write a Python program to visualize the F-distribution and discuss its use in hypothesis testing.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import f

# Step 1: Define degrees of freedom
dfn = 5   # Degrees of freedom numerator (between-group)
dfd = 20  # Degrees of freedom denominator (within-group)

# Step 2: Generate F-distribution values
x = np.linspace(0, 5, 500)
y = f.pdf(x, dfn, dfd)

# Step 3: Critical value for 95% confidence
alpha = 0.05
f_critical = f.ppf(1 - alpha, dfn, dfd)

# Step 4: Plot F-distribution
plt.figure(figsize=(8, 5))
plt.plot(x, y, 'b-', label=f'F-distribution (dfn={dfn}, dfd={dfd})')
plt.axvline(f_critical, color='r', linestyle='--', label=f'Critical value = {f_critical:.2f}')
plt.fill_between(x, y, where=(x > f_critical), color='red', alpha=0.3, label='Rejection region')

plt.title('F-Distribution with Rejection Region')
plt.xlabel('F-value')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#Q23. Simulate random data from a normal distribution, then perform hypothesis testing to evaluate the means.

import numpy as np
from scipy import stats

# Step 1: Simulate random data from normal distribution
np.random.seed(42)
sample = np.random.normal(loc=50, scale=10, size=30)  # Sample from N(50, 10¬≤)

# Step 2: Set population mean (¬µ0) to test against
pop_mean = 52  # H‚ÇÄ: sample mean = 52

# Step 3: Perform one-sample T-test
t_stat, p_val = stats.ttest_1samp(sample, pop_mean)

# Step 4: Display results
print("One-Sample T-Test Results")
print(f"Sample Mean = {np.mean(sample):.2f}")
print(f"T-statistic = {t_stat:.4f}")
print(f"P-value = {p_val:.4f}")

alpha = 0.05
if p_val < alpha:
    print("Conclusion: Reject H‚ÇÄ ‚Äî Sample mean is significantly different from population mean.")
else:
    print("Conclusion: Fail to reject H‚ÇÄ ‚Äî No significant difference from population mean.")
```

```python
#Q24. Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results.

import numpy as np
from scipy.stats import chi2

# Step 1: Simulate sample data
np.random.seed(42)
sample = np.random.normal(loc=50, scale=10, size=30)

# Step 2: Define hypothesized variance
sigma_0_sq = 100  # Hypothesized variance (e.g., std dev = 10)

# Step 3: Compute sample variance
n = len(sample)
sample_var = np.var(sample, ddof=1)  # unbiased estimate (ddof=1)

# Step 4: Compute Chi-square statistic
chi2_stat = (n - 1) * sample_var / sigma_0_sq

# Step 5: Compute p-value (two-tailed test)
alpha = 0.05
p_left = chi2.cdf(chi2_stat, df=n-1)
p_right = 1 - chi2.cdf(chi2_stat, df=n-1)
p_value = 2 * min(p_left, p_right)

# Step 6: Critical values (for visualization or rejection region)
lower_crit = chi2.ppf(alpha / 2, df=n - 1)
upper_crit = chi2.ppf(1 - alpha / 2, df=n - 1)

# Step 7: Print results
print("Chi-Square Test for Population Variance")
print(f"Sample Variance = {sample_var:.2f}")
print(f"Chi-square Statistic = {chi2_stat:.4f}")
print(f"P-value = {p_value:.4f}")
print(f"Chi-square Critical Values (Œ±=0.05): [{lower_crit:.2f}, {upper_crit:.2f}]")

if chi2_stat < lower_crit or chi2_stat > upper_crit:
    print("Conclusion: Reject H‚ÇÄ ‚Äî Population variance is significantly different from hypothesized.")
else:
    print("Conclusion: Fail to reject H‚ÇÄ ‚Äî No significant difference in variance.")
```

```python
#Q26. Implement an F-test for comparing the variances of two datasets, then interpret and visualize the results.

import numpy as np
from scipy.stats import f
import matplotlib.pyplot as plt

# Step 1: Simulate two datasets
np.random.seed(42)
data1 = np.random.normal(loc=50, scale=10, size=30)  # std dev = 10
data2 = np.random.normal(loc=50, scale=15, size=30)  # std dev = 15

# Step 2: Calculate sample variances
var1 = np.var(data1, ddof=1)
var2 = np.var(data2, ddof=1)

# Step 3: Calculate F-statistic (larger variance / smaller variance)
f_stat = var1 / var2 if var1 > var2 else var2 / var1
df1 = len(data1) - 1
df2 = len(data2) - 1

# Step 4: Calculate two-tailed p-value
p_val = 2 * min(f.cdf(f_stat, df1, df2), 1 - f.cdf(f_stat, df1, df2))

# Step 5: Critical values for Œ± = 0.05
alpha = 0.05
f_crit_low = f.ppf(alpha / 2, df1, df2)
f_crit_high = f.ppf(1 - alpha / 2, df1, df2)

# Step 6: Print results
print("F-Test for Equality of Variances")
print(f"Variance 1: {var1:.2f}, Variance 2: {var2:.2f}")
print(f"F-statistic: {f_stat:.4f}")
print(f"P-value: {p_val:.4f}")
print(f"Critical values (Œ± = 0.05): [{f_crit_low:.2f}, {f_crit_high:.2f}]")

if f_stat < f_crit_low or f_stat > f_crit_high:
    print("Conclusion: Reject H‚ÇÄ ‚Äî Variances are significantly different.")
else:
    print("Conclusion: Fail to reject H‚ÇÄ ‚Äî No significant difference in variances.")

# Step 7: Visualize the F-distribution
x = np.linspace(0.1, 5, 500)
y = f.pdf(x, df1, df2)

plt.figure(figsize=(8, 5))
plt.plot(x, y, 'b-', label='F-distribution')
plt.axvline(f_stat, color='red', linestyle='--', label=f'F-stat = {f_stat:.2f}')
plt.axvline(f_crit_low, color='black', linestyle=':', label='Lower Crit. Value')
plt.axvline(f_crit_high, color='black', linestyle=':', label='Upper Crit. Value')
plt.fill_between(x, y, where=(x < f_crit_low) | (x > f_crit_high), color='red', alpha=0.3, label='Rejection Region')

plt.title('F-Test: Comparison of Two Variances')
plt.xlabel('F-value')
plt.ylabel('Density')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#Q27.  Perform a Chi-square test for goodness of fit with simulated data and analyze the results.

import numpy as np
from scipy.stats import chisquare
import matplotlib.pyplot as plt

# Step 1: Simulate observed data (e.g., rolling a biased die)
np.random.seed(42)
observed = np.random.choice([1, 2, 3, 4, 5, 6], size=120, p=[0.10, 0.20, 0.15, 0.15, 0.20, 0.20])
observed_counts = np.bincount(observed)[1:]  # Skip index 0

# Step 2: Define expected distribution (uniform die)
expected_counts = np.array([20, 20, 20, 20, 20, 20])  # Equal frequency for 6 sides

# Step 3: Perform Chi-square goodness-of-fit test
chi_stat, p_val = chisquare(f_obs=observed_counts, f_exp=expected_counts)

# Step 4: Print results
print("Chi-Square Goodness-of-Fit Test")
print(f"Observed Counts: {observed_counts}")
print(f"Expected Counts: {expected_counts}")
print(f"Chi-square Statistic: {chi_stat:.4f}")
print(f"P-value: {p_val:.4f}")

alpha = 0.05
if p_val < alpha:
    print("Conclusion: Reject H‚ÇÄ ‚Äî The observed distribution differs from the expected.")
else:
    print("Conclusion: Fail to reject H‚ÇÄ ‚Äî No significant difference from the expected distribution.")

# Step 5: Visualize the results
categories = ['1', '2', '3', '4', '5', '6']
x = np.arange(len(categories))

plt.bar(x - 0.2, observed_counts, width=0.4, label='Observed', color='skyblue')
plt.bar(x + 0.2, expected_counts, width=0.4, label='Expected', color='orange')
plt.xticks(x, categories)
plt.title('Chi-Square Goodness-of-Fit: Observed vs Expected')
plt.xlabel('Category')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
